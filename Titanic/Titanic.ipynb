Cek keberadaan data null dan tipe data dengan menggunakan .info(). Jika terdapat nilai null ada beberapa hal yang bisa dilakukan:
1. Jika nilai null ada pada kolom dengan tipe data numerik bisa diganti dengan nilai mean atau median
2. Jika nilai null ada pada kolom dengan tipe data string atau object dapat dihapus semua baris yang bersangkutan
3. Jika ada banyak nilai null pada kolom tersebut dan dirasa informasi yang diberikan dari kolom tersebut tidak bermanfaat banyak di dalam pembuatan model maka kolom tersebut dapat dihapus.
4. Jika ada banyak nilai null pada kolom tersebut dan dirasa informasi yang diberikan dari kolom tersebut bermanfaat banyak di dalam pembuatan model maka filter dataframe sesuai dengan data yang lengkap

```python
median = dataset['age'].median()
dataset['age'].fillna(median, inplace = True)
median
```

```python
import numpy as np

for i in range(0,len(dataset)):
    if 0 < dataset['age'][i] <= 12:
        dataset['age'][i] = 5
    elif 12 < dataset['age'][i] <= 18:
        dataset['age'][i] = 4
    elif 18 < dataset['age'][i] <= 35:
        dataset['age'][i] = 3 
    elif 35 < dataset['age'][i] <= 64:
        dataset['age'][i] = 2 
    elif 64 < dataset['age'][i] <= np.inf:
        dataset['age'][i] = 1
```

```python
dataset['age'] = dataset['age'].astype('int64')
```

```python
mean = dataset['fare'].mean()
dataset['fare'].fillna(mean, inplace = True)
mean
```

```python
for i in range(0,len(dataset)):
    if 0 < dataset['fare'][i] <= 10:
        dataset['fare'][i] = 1
    elif 10 < dataset['fare'][i] <= 50:
        dataset['fare'][i] = 2
    elif 50 < dataset['fare'][i] <= 100:
        dataset['fare'][i] = 3 
    elif 100 < dataset['fare'][i] <= 200:
        dataset['fare'][i] = 4 
    elif 200 < dataset['fare'][i] <= np.inf:
        dataset['fare'][i] = 5
```

```python
dataset['fare'] = dataset['fare'].astype('int64')
```

```python
dataset['embarked'].fillna('S', inplace = True)
```

```python
dataset = dataset.drop(['name','ticket','cabin','boat','body','home.dest'], axis=1)
```

```python
dataset = dataset.dropna()
```

```python
embarked_map = {'S':0, "C":1, "Q":2}
dataset['embarked'] = dataset['embarked'].map(embarked_map)
```

```python
dataset.info()
```

Jika ada data yang bersifat kategorikal dapat dilakukan OneHotEncoder

```python
from sklearn.preprocessing import OneHotEncoder
OHE=OneHotEncoder()
encode_sex = OHE.fit_transform(dataset[['sex']])
dataset['sex'] = encode_sex.toarray().astype('int64')
```

Bagi dataset menjadi data latih dan data uji dengan perbandingan data latih 80% dan data uji 20%. Pastikan proporsi data latih untuk penumpang yang merata, gunakan StratifiedShuffleSplit.

```python
from sklearn.model_selection import StratifiedShuffleSplit
split = StratifiedShuffleSplit(n_splits =1, test_size=0.2, random_state=42)
split
```

```python
for train_index, test_index in split.split(dataset, dataset['sex']):
    strat_train_set = dataset.loc[train_index]
    strat_test_set = dataset.loc[test_index]
```

```python
dataset_features = strat_train_set.drop(['survived'], axis=1)
dataset_labels = strat_train_set['survived'].copy()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(dataset_features, dataset_labels, test_size = 0.2, random_state = 0)
```

Lakukan feature scaling 

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

Buat model pembelajaran mesin untuk klasifikasi dengan menggunakan algoritma yang sudah diajarkan: Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier

classifier = DecisionTreeClassifier(max_depth = 2)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
y_pred
```

Buat confusion matrix untuk masing-masing algoritma

```python
from sklearn.metrics import confusion_matrix
confu_matrix = confusion_matrix(y_test, y_pred)
confu_matrix
```

Tampilkan keduanya di dalam kurva ROC dan Precision/Recall lalu tentukan algoritma mana yang lebih baik

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score


lr_auc = roc_auc_score(y_test, y_pred)
print('Logistic: ROC AUC=%.3f' % (lr_auc))
lr_fpr, lr_tpr, _ = roc_curve(y_test, y_pred)
plt.plot(lr_fpr, lr_tpr, marker='.', label='Logistic')
plt.xlabel('False Positive')
plt.ylabel('True Positive')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import precision_recall_curve

lr_fpr, lr_tpr, _ = precision_recall_curve(y_test, y_pred)
plt.plot(lr_fpr, lr_tpr, marker='.', label='Logistic')
plt.xlabel('False Positive')
plt.ylabel('True Positive')
plt.legend()
plt.show()
```
