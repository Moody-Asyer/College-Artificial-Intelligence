## Panggil Dataset

```python
# scatter plot of moons dataset
from sklearn.datasets import make_moons
from matplotlib import pyplot
from numpy import where
# generate 2d classification dataset
X, y = make_moons(n_samples=100, noise=0.2, random_state=1)
# scatter plot for each class value
for class_value in range(2):
    # select indices of points with the class label
    row_ix = where(y == class_value)
    # scatter plot for points with a different color
    pyplot.scatter(X[row_ix, 0], X[row_ix, 1])
# show plot
pyplot.show()
```

## Latih Model Tanpa Early Stopping

```python
# mlp overfit on the moons dataset
from keras import models
from keras import layers
from matplotlib import pyplot

# split into train and test
n_train = 30
trainX, testX = X[:n_train, :], X[n_train:, :]
trainy, testy = y[:n_train], y[n_train:]
# define model
model = models.Sequential()
model.add(layers.Dense(500, input_dim=2, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.summary()
```

Perhatikan, apa yang terjadi ketika kita menentukan epochs terlalu besar?

```python
# fit model
history = model.fit(trainX, trainy, validation_data=(testX, testy), epochs=4000)
```

Dari mana kita tahu terjadi overfitting saat epochs terlalu besar? Salah satunya bisa dilihat pada saat adanya titik balik pada validation loss, dari awalnya bergerak ke bawah kemudian kembali naik nilainya. Lalu bisa juga dilihat dari validation accuracy yang awalnya naik bergerak menuju 1, tetapi pada satu titik dia mulai menurun.

Secara pengamatan kita tahu, jika harus menerapkan early stopping, maka epochs yang harus digunakan ada di sekitar 800.

```python
# evaluate the model
_, train_acc = model.evaluate(trainX, trainy, verbose=0)
_, test_acc = model.evaluate(testX, testy, verbose=0)
print('Train: %.3f, Test: %.3f' % (train_acc, test_acc))
# plot loss learning curves
pyplot.subplot(211)
pyplot.title('Cross-Entropy Loss', pad=-40)
pyplot.plot(history.history['loss'], label='train')
pyplot.plot(history.history['val_loss'], label='test')
pyplot.legend()
# plot accuracy learning curves
pyplot.subplot(212)
pyplot.title('Accuracy', pad=-40)
pyplot.plot(history.history['accuracy'], label='train')
pyplot.plot(history.history['val_accuracy'], label='test')
pyplot.legend()
pyplot.show()
```

## Latih model dengan early stopping
Definisikan kembali model yang sama dengan yang di atas

```python
# define model
model = models.Sequential()
model.add(layers.Dense(500, input_dim=2, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

Definisikan early stopping. Verbose 1 berarti kita ingin mengetahui di epoch ke berapa proses pelatihan akan distop.

```python
from keras import callbacks
# simple early stopping
es = callbacks.EarlyStopping(monitor='val_loss', mode='min', verbose=1)
```

Tambahkan early stopping pada saat kita akan melakukan pembuatan model.

```python
# fit model
history = model.fit(trainX, trainy, validation_data=(testX, testy), epochs=4000, verbose=1, callbacks=[es])
```

Dari hasil di atas kita bisa lihat kalau proses early stopping berhenti di epoch ke 219. Padahal secara intuisi kita tahu seharusnya dia berhenti di sekitaran epoch ke 800. Hal ini juga dipertegas dengan hasil akurasi dari klasifikasi menghasilkan hasil yang lebih rendah dibandingkan yang tidak menggunakan early stopping (lihat validation accuracy/test accuracy)

```python
# evaluate the model
_, train_acc = model.evaluate(trainX, trainy, verbose=0)
_, test_acc = model.evaluate(testX, testy, verbose=0)
print('Train: %.3f, Test: %.3f' % (train_acc, test_acc))
# plot loss learning curves
pyplot.subplot(211)
pyplot.title('Cross-Entropy Loss', pad=-40)
pyplot.plot(history.history['loss'], label='train')
pyplot.plot(history.history['val_loss'], label='test')
pyplot.legend()
# plot accuracy learning curves
pyplot.subplot(212)
pyplot.title('Accuracy', pad=-40)
pyplot.plot(history.history['accuracy'], label='train')
pyplot.plot(history.history['val_accuracy'], label='test')
pyplot.legend()
pyplot.show()
```

Untuk meningkatkan kualitas dari early stopping yang kita miliki bisa dilakukan dengan menambahkan argumen patience. Di dalam kasus ini, kita akan menunggu 200 epochs sebelum pelatihan dihentikan. Secara spesifik, hal ini berarti kita akan memberikan tambahan 200 epochs setelah terjadi perubahan arah validation loss untuk memastikan early stopping menghentikan proses pelatihan di epochs yang tepat.

**Definisikan model kembali**

```python
# define model
model = models.Sequential()
model.add(layers.Dense(500, input_dim=2, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# patience simple early stopping
es = callbacks.EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=200)
```

```python
# fit model
history = model.fit(trainX, trainy, validation_data=(testX, testy), epochs=4000, verbose=1, callbacks=[es])
```

Hasil di bawah memberikan kita hasil yang lebih baik, di mana model berhenti di epochs sekitar 800an, yaitu 964. Selain itu, nilai test accuracy juga lebih tinggi dibandingkan metode early stopping sebelum ini.

```python
# evaluate the model
_, train_acc = model.evaluate(trainX, trainy, verbose=0)
_, test_acc = model.evaluate(testX, testy, verbose=0)
print('Train: %.3f, Test: %.3f' % (train_acc, test_acc))
# plot loss learning curves
pyplot.subplot(211)
pyplot.title('Cross-Entropy Loss', pad=-40)
pyplot.plot(history.history['loss'], label='train')
pyplot.plot(history.history['val_loss'], label='test')
pyplot.legend()
# plot accuracy learning curves
pyplot.subplot(212)
pyplot.title('Accuracy', pad=-40)
pyplot.plot(history.history['accuracy'], label='train')
pyplot.plot(history.history['val_accuracy'], label='test')
pyplot.legend()
pyplot.show()
```

## Membuat Checkpoint
Meskipun penambahan patience sudah menghasilkan akurasi testing yang lebih baik, namun bisa dilihat kalau di sekitar 100 epochs terakhir nilai validation loss/test loss sudah mulai naik kembali. Hal ini berarti meskipun performansi dari model sudah meningkat, kita masih belum memiliki model yang paling baik. Kita dapat menyelesaikan permasalah ini dengan menggunakan **ModelCheckpoint** callback. Di dalam kasus ini kita ingin menyimpan model dengan nilai akurasi paling tinggi.

```python
# define model
model = models.Sequential()
model.add(layers.Dense(500, input_dim=2, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# add checkpoint
mc = callbacks.ModelCheckpoint('best_model.h5', monitor='val_accuracy', mode='max', verbose=1, save_best_only=True)
```

Tambahkan model checkpoint pada saat melatih model. Perhatikan di epoch-epoch tertentu model akan menyimpan model yang memiliki performa paling baik. Meskipun early stopping berhenti di epoch ke-968, namun model yang di save adalah model di epoch ke-848.

```python
# fit model
history = model.fit(trainX, trainy, validation_data=(testX, testy), epochs=4000, verbose=0, callbacks=[es, mc])
```

Untuk membuktikannya kita akan memanggil kembali model dengan nama 'best_model.h5' yang sebelumnya sudah disimpan dan melihat berapa nilai akurasi model tersebut.

```python
from keras import models
# load the saved model
saved_model = models.load_model('best_model.h5')
# evaluate the model
_, train_acc = saved_model.evaluate(trainX, trainy, verbose=0)
_, test_acc = saved_model.evaluate(testX, testy, verbose=0)
print('Train: %.3f, Test: %.3f' % (train_acc, test_acc))
```
